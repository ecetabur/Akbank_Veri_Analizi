```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris = pd.read_csv(url, names=columns)

# Veri setinin ilk 5 satırı
print("Veri setinin ilk 5 satırı:")
print(iris.head())

# Eksik veri kontrolü
print("\nEksik veri kontrolü:")
print(iris.isnull().sum())

# Temel istatistiksel analizler
print("\nTemel İstatistiksel Analizler:")
print(iris.describe())

# Görselleştirme
# Çiçek türlerine göre çanak yaprak uzunluklarının dağılımı
sns.displot(data=iris, x='sepal_length', hue='class', kde=True)
plt.title('Çiçek Türlerine Göre Sepal Uzunluklarının Dağılımı')
plt.xlabel('Sepal Uzunluğu (cm)')
plt.show()

# Çiçek türlerine göre taç yaprak uzunlukları ve genişliklerinin ilişkisi
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='class')
plt.title('Çiçek Türlerine Göre Petal Uzunluğu ve Genişliği İlişkisi')
plt.xlabel('Petal Uzunluğu (cm)')
plt.ylabel('Petal Genişliği (cm)')
plt.show()
```
